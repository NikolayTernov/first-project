````python
def df_starter(df):
    df_na = df.isna().mean() * 100
    styled_df_na = df_na.round(1).to_frame().style.background_gradient("coolwarm")
    df_dup = df.duplicated().sum()
    
    display(styled_df_na, f'Количество дубликатов по столбцам: {df_dup}')
````

Сделаем гистограммы для каждой группы по целевому признаку 

````python
for col in market_file.select_dtypes(exclude='object').columns:
    active_customers = market_file[market_file['Покупательская активность'] == "Снизилась"][col]
    inactive_customers = market_file[market_file['Покупательская активность'] == "Прежний уровень"][col]
    plt.figure(figsize=(8, 6))
    plt.hist(active_customers, alpha=0.5, label='Юзеры с прежнем уровнем ак-ти')
    plt.hist(inactive_customers, alpha=0.5, label='Юзеры со снизившемся уровнем ак-ти')
    plt.xlabel(f'Значения в колонке {col}')
    plt.ylabel('Частота')
    plt.title(f'Распределение в колонке "{col}"  для двух категорий покупателей')
    plt.legend()
    plt.show()
    
    
````

3. ADASYN - синтез данных "на грани", полезно когда классы находятся рядом. В нашем случае это пока не определить;
4. Андерсемплинг - логичное решение, но данных не так много, и не хотелось бы жертвовать ими. Было бы 8к против 4к - не вопрос, а так - прибережем;
5. Tomek Links - компромисс, но он плохо работает с пограничными решениями;
6. `class_weight='balanced'` из sklearn - попытка задать "веса", но не поддерживается в KNN-модели.

3. На основе данных о выручке из пункта 1 можно посмотреть на долю акционных покупок и выявить акционную активность;

Начнем работу с пункта 1. 

````python
df_decreased['Итог выручки'] = ''

# Сравнение выручки за каждый месяц и присвоение значения в колонку "итог"
for index, row in df_decreased.iterrows():
    if row['выручка_предыдущ_мес'] < row['выручка_предпред_мес'] and row['выручка_текущ_мес'] < row['выручка_предыдущ_мес']:
        df_decreased.at[index, 'Итог выручки'] = 'снизилась'
    elif row['выручка_предыдущ_мес'] > row['выручка_предпред_мес'] and row['выручка_текущ_мес'] > row['выручка_предыдущ_мес']:
        df_decreased.at[index, 'Итог выручки'] = 'повысилась'
    else:
        df_decreased.at[index, 'Итог выручки'] = 'прежняя'
````

# ОПЦИОНАЛЬНО - КЛАСТЕРИЗАЦИЯ #

Вообще, сегментация (кластеризация) - вопрос сложный. В ML есть много методов, среди которых выделяют K-Means. Тут я просто хочу попробовать сегментировать юзеров, чья активность снизилась и посмотреть, что вышло.

В учебниках говорится, что в случае, если датафрейм имеет много данных, есть смысл реализовать PCA и посмотреть, какие 2-3 группы присутствуют. Сделаем это.

<div class="alert alert-success">
<h2> Комментарий ревьюера ✔️ <a class="tocSkip"> </h2>

Дополнительно про кластеризацию можно почитать вот [тут](https://academy.yandex.ru/handbook/ml/article/klasterizaciya) - по мне, лучшее что есть на РУ

````python
from sklearn.decomposition import PCA
````

````python
df_decr_dummies = df_decreased.drop(['Итог выручки', "Доля ошибок", "Итог маркет_актив", 'id', 'Покупательская активность'], axis = 1)
df_decr_dummies = pd.get_dummies(df_decr_dummies, columns=['Тип сервиса', 'Разрешить сообщать', "Популярная_категория"])
````

Суть метода - все данные можно "сжать" до 2 компонентов, тем самым, уменьшая размерность. Запилим компоненты и посмотрим, какие у нас 2 главных компонента.

````python
pca = PCA(n_components=2)

pca_components = pca.fit_transform(df_decr_dummies.values)
pca.explained_variance_ratio_
````

Как я понял, первый компонент "ответственен" за 95% инфы, второй - всего за 3%. Посмотрим, какие две группы "сформированы" среди нашего сегмента (чья активность снизилась)

````python
df_pca = pd.DataFrame(pca.components_,columns=df_decr_dummies.columns,index = ['PC-1','PC-2'])
df_pca.T
````

С  помощью РСА можно теперь реализовать классический алгоритм - K-Means.  Пусть кластеров будет 3

<div class="alert alert-success">
<h2> Комментарий ревьюера ✔️ <a class="tocSkip"> </h2>
    
Можешь почитать про иерархическую кластеризацию, про методы  linkage() и дендрограмму...

````python
from sklearn.cluster import KMeans


kmeans = KMeans(n_clusters=3, random_state=RANDOM_STATE)
kmeans.fit(pca_components)
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

````

Делаем график кластеров и РСА. Как я понял, в итоге у нас их реально 3, причем 1 даже "лишний" - он справа. Получается, всю группу можно поделить на две категории?

<div class="alert alert-success">
<h2> Комментарий ревьюера ✔️ <a class="tocSkip"> </h2>

Это говорит, что их СКОРЕЕ ВСЕГО 3, но может быть и 2 и 3 и 4..., где то в райное трех)

````python
plt.figure(figsize=(6, 4))
plt.scatter(pca_components[:, 0], pca_components[:, 1], c=cluster_labels, cmap='viridis', marker='o', edgecolor='black')

plt.xlabel('Главный компонент 1')
plt.ylabel('Главный компонент 2')

plt.colorbar(label='Cluster')

plt.title('Визуализация после PCA и К-Means')
plt.show()
````

<div class="alert alert-success">

Посмотрим как расположены кластеры на графиках

````python
plt.figure(figsize=(6, 3))
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', s=100, c='red', label='Центры кластеров')

plt.legend()

plt.xlabel('Главный компонент 1')
plt.ylabel('Главный компонент 2')
plt.title('Центры кластеров')

plt.show()
````

Да, у нас три кластера, но те, что слева все же "близки" друг к другу. 
Сделаем датафреймы из наших кластеров. 

````python
df_with_clusters = df_decr_dummies.copy()
df_with_clusters['Cluster'] = cluster_labels
````

````python
cluster_0_df = df_with_clusters[df_with_clusters['Cluster'] == 0].drop('Cluster', axis=1)
cluster_1_df = df_with_clusters[df_with_clusters['Cluster'] == 1].drop('Cluster', axis=1)
cluster_2_df = df_with_clusters[df_with_clusters['Cluster'] == 2].drop('Cluster', axis=1)
````

Посмотрим на датафреймы. Предлагаю сделать 1 датафрейм с медианными значениями по каждой категории

````python
dataframes = ['cluster_0_df', 'cluster_1_df', 'cluster_2_df']
dfs = [cluster_0_df, cluster_1_df, cluster_2_df]
medians = []

for df in dfs:
    median_values = df.median()
    medians.append(median_values)

result_df = pd.concat(medians, axis=1)
result_df.columns = dataframes 
````
